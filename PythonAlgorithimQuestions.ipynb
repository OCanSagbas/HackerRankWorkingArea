```python
"""Sample Input 0
HackerRank.com presents "Pythonist 2".
Sample Output 0
hACKERrANK.COM PRESENTS "pYTHONIST 2".

Name : Swap Case
"""
b=[]
def swap_case(s):
    for a in s:
        if isinstance(a, str) and a.isupper():
            a = a.lower()
        elif isinstance(a, str) and a.islower():
            a = a.upper()
        b.append(a)
    return "".join(b)
if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
"""
Determining the minimum time required to process files using multiple cores, with constraints on the number of parallelizable tasks.
Name: Parallel Processing
"""
def minTime(files, numCores, limit):
    # Separate files into those that can and cannot be processed in parallel
    parallelizable = [file for file in files if file % numCores == 0]
    non_parallelizable = [file for file in files if file % numCores != 0]

    # Sort parallelizable files in descending order for optimal usage of the limit
    parallelizable.sort(reverse=True)

    # Initialize total processing time
    total_time = sum(non_parallelizable)  # Non-parallelizable tasks are processed sequentially

    # Process parallelizable files within the limit
    for i in range(min(limit, len(parallelizable))):
        total_time += parallelizable[i] // numCores

    # Process remaining parallelizable files sequentially
    for i in range(min(limit, len(parallelizable)), len(parallelizable)):
        total_time += parallelizable[i]

    return total_time

# Example usage in HackerRank's input-output structure
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    files_count = int(input().strip())

    files = []

    for _ in range(files_count):
        files_item = int(input().strip())
        files.append(files_item)

    numCores = int(input().strip())

    limit = int(input().strip())

    result = minTime(files, numCores, limit)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#!/bin/python3
"""
Name: The Power Sum
Example = X:13, N:2
13 = 2^2+3^2
Output:1
"""
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'powerSum' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER X
#  2. INTEGER N
#
def powerSum(x, n, num=1):
    # Base cases
    val = (x - pow(num, n))
    if (val == 0):
        return 1
    if (val < 0):
        return 0

    # Consider two possibilities, num is
    # included and num is not included.
    return powerSum(val, n, num + 1) + powerSum(x, n, num + 1)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    X = int(input().strip())

    N = int(input().strip())

    result = powerSum(X, N)

    fptr.write(str(result) + '\n')

    fptr.close()
```
